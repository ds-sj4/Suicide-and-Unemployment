```python
#Africa Eastern and Southern Regression
from sklearn.linear_model import LinearRegression

est = LinearRegression(fit_intercept = True)

x = df_aes[['Unemployment, total (% of total labor force)']]
y = df_aes[['Suicide mortality rate (per 100,000 population)']]

est.fit(x, y)

print("Coefficients:", est.coef_)
print ("Intercept:", est.intercept_)
```
